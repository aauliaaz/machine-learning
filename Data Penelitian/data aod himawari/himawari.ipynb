```python
def download_nc_files(ftp, files, local_dir):
    if not os.path.exists(local_dir):
        os.makedirs(local_dir)

    for file_name in files:
        if file_name.endswith('.nc'):
            local_file_path = os.path.join(local_dir, file_name)
            local_size = os.path.getsize(local_file_path) if os.path.exists(local_file_path) else 0
            
            try:
                ftp.voidcmd("TYPE I")  # Mode biner
                remote_size = ftp.size(file_name)
            except Exception:
                remote_size = None
                print(f"Tidak bisa mendapatkan ukuran {file_name}, mulai ulang download.")
            
            if remote_size and local_size == remote_size:
                print(f"File {file_name} sudah ada, skip download.")
                continue
            
            with open(local_file_path, 'ab') as f:
                ftp.retrbinary(f'RETR {file_name}', f.write, rest=local_size)
            print(f"Selesai download: {local_file_path}")
```

```python
import datetime

def download_himawari_data(ftp_user, ftp_password, start_year, start_month, end_year, end_month, local_base_dir):
    ftp = FTP("ftp.ptree.jaxa.jp")
    ftp.login(ftp_user, ftp_password)
    print("Logged in to FTP server.")

    start_date = datetime.date(start_year, start_month, 1)
    end_date = datetime.date(end_year, end_month, 1)
    current_date = start_date
    
    while current_date <= end_date:
        year = current_date.year
        month = f"{current_date.month:02d}"  # Format dua digit
        remote_dir = f"/pub/himawari/L3/ARP/031/{year}{month}/daily"
        # local_dir = os.path.join(local_base_dir, f"{year}_{month}")
        local_dir = local_base_dir
        
        try:
            ftp.cwd(remote_dir)
            files = sorted(ftp.nlst())
            print(f"Mengunduh data dari: {remote_dir}")
            download_nc_files(ftp, files, local_dir)
        except Exception as e:
            print(f"Gagal mengakses {remote_dir}: {e}")
        
        # Pindah ke bulan berikutnya
        if current_date.month == 12:
            current_date = datetime.date(current_date.year + 1, 1, 1)
        else:
            current_date = datetime.date(current_date.year, current_date.month + 1, 1)
    
    ftp.quit()
    print("Selesai mengunduh semua data.")
```

```python
# Batas wilayah Jakarta
import os
min_lon, max_lon = 106.5, 107.0
min_lat, max_lat = -6.5, -5.9

def convert_nc_to_csv_jakarta(nc_file, output_dir="C:/AData/kuliah/sem 7/Tugas Akhir/Kode Program/data aod himawari/tes csv"):
    """Membaca file .nc, memfilter hanya data AOD di Jakarta, dan menyimpannya dalam format .csv"""
    
    # Pastikan direktori output ada
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # Buka file NetCDF
    ds = xr.open_dataset(nc_file)

    # Cek nama variabel longitude & latitude
    lon_name = [name for name in ds.coords if 'lon' in name.lower()][0]  # Cari otomatis
    lat_name = [name for name in ds.coords if 'lat' in name.lower()][0]
    
    # Filter untuk wilayah Jakarta (sekitar 106.7E, -6.2S)
    jakarta_data = ds.where(
        (ds[lon_name] >= min_lon) & (ds[lon_name] <= max_lon) & 
        (ds[lat_name] >= min_lat) & (ds[lat_name] <= max_lat),
        drop=True
    )
    
    # Konversi ke DataFrame
    df = jakarta_data.to_dataframe().reset_index()
    
    # Nama file CSV
    csv_file = os.path.join(output_dir, os.path.basename(nc_file).replace(".nc", "_jakarta.csv"))
    
    # Simpan sebagai CSV
    df.to_csv(csv_file, index=False)
    
    print(f"Converted (Jakarta only): {nc_file} → {csv_file}")
```

```python
def convert_nc_to_csv(nc_file_path, output_folder):
    # Baca file NetCDF
    ds = xr.open_dataset(nc_file_path)
    
    # Tentukan batas koordinat Jakarta
    min_lon, max_lon = 106.5, 107.0
    min_lat, max_lat = -6.5, -5.9
    
    # Ekstrak variabel yang relevan
    df = ds.to_dataframe().reset_index()
    
    # Filter berdasarkan koordinat Jakarta
    df_jakarta = df[(df['longitude'] >= min_lon) & (df['longitude'] <= max_lon) &
                    (df['latitude'] >= min_lat) & (df['latitude'] <= max_lat)]
    
    # Dapatkan nama file tanpa ekstensi
    base_name = os.path.splitext(os.path.basename(nc_file_path))[0]
    
    # Simpan sebagai CSV
    output_csv_path = os.path.join(output_folder, f"{base_name}.csv")
    df_jakarta.to_csv(output_csv_path, index=False)
    
    print(f"File berhasil dikonversi: {output_csv_path}")
```

```python
nc_folder = 'C:/AData/kuliah/sem 7/Tugas Akhir/Kode Program/data aod himawari/2024'
output_dir = 'C:/AData/kuliah/sem 7/Tugas Akhir/Kode Program/data aod himawari/2024 csv'

for file in os.listdir(nc_folder):
    if file.endswith('.nc'):
        convert_nc_to_csv_jakarta(os.path.join(nc_folder, file),output_dir)
        #print(file)
```
